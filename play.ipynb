```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Create a tensor with input values
x = torch.linspace(-5, 5, 100)

# Initialize the nn.GELU activation function
gelu = nn.GELU()

# Compute GELU for each input value
gelu_values = gelu(x)

# Plot the GELU function
plt.plot(x, gelu_values)
plt.xlabel('Input')
plt.ylabel('nn.GELU(x)')
plt.title('Plot of nn.GELU Activation Function')
plt.grid(True)
plt.show()
```
